```python
import pandas as pd
import quantstats as qs
```

```python
spy = pd.read_csv("./charts/spy_historical.csv")
spy
```

```python
spy['year'] = spy['Date'].apply(lambda x: x[-4:])
spy
```

```python
spy_2018 = spy[spy['year'] == '2018']
spy_2018
```

```python
spy_close = pd.to_numeric(spy_2018['Close/Last'])
```

```python
ret = spy_close.diff().to_numpy()
ret = ret[1:] / spy_close.to_numpy()[:-1]
ret
```

```python
qs.stats.sharpe(pd.Series(ret))
```
