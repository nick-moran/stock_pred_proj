```python
import os
import json

def filter_articles(news_dir):
  news_files = os.listdir(news_dir)

  news_info = []
  for file in news_files:
      with open(f'./{news_dir}/{file}') as article:
        article_content = json.load(article)
        aapl = -1
        amzn = -1
        for org in article_content['entities']['organizations']:
          if ('apple' in article_content['text'].lower()) and aapl < 0:
            news_info.append(['aapl', article_content['published'], article_content['text'], f'{news_dir}/{file}'])
            aapl *= -1

          if ('amazon' in article_content['text'].lower()) and amzn < 0:
            news_info.append(['amzn', article_content['published'], article_content['text'], f'{news_dir}/{file}'])
            amzn *= -1
  return news_info
```

```python
def filter_articles(news_dir):
  news_files = os.listdir(news_dir)

  news_info = []
  for file in news_files:
      with open(f'./{news_dir}/{file}') as article:
        article_content = json.load(article)
        aapl = -1
        amzn = -1
        for org in article_content['entities']['organizations']:
          if ('apple' in article_content['text'].lower()) and aapl < 0:
            news_info.append(['aapl', article_content['published'], article_content['text'], f'{news_dir}/{file}'])
            aapl *= -1

          if ('amazon' in article_content['text'].lower()) and amzn < 0:
            news_info.append(['amzn', article_content['published'], article_content['text'], f'{news_dir}/{file}'])
            amzn *= -1
  return news_info
```

```python
news_dirs = os.listdir('./news')

ndf= []
for news_dir in news_dirs:
    ndf += filter_articles(f'./news/{news_dir}')

print(len(ndf))
```

```python
import pandas as pd
news_data = pd.DataFrame(ndf, columns=['company', 'date','text','file'])
news_data
```

```python
from textblob import TextBlob

def get_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    if sentiment > 0:
        return 1
    elif sentiment < 0:
        return -1
    else:
        return 0

# add new column with sentiment
news_data['sentiment'] = news_data['text'].apply(get_sentiment)

news_data
```

```python
news_data['date'] = pd.to_datetime(news_data['date'], utc=True)
news_dates = news_data['date'].apply(lambda x: x.strftime('%Y.%m.%d'))
news_times = news_data['date'].apply(lambda x: x.strftime('%H:%M'))

news_data['date'] = news_dates
news_data['time'] = news_times
```

```python
apple_news = news_data[news_data['company'] == 'aapl']
apple_news
```

```python
news_df = pd.read_csv('/content/senti/all-data.csv', encoding = "ISO-8859-1", names = ['lables','messages'])
news_df.head()
```

```python
df_copy = news_df.copy()
df_copy.head()
def func(df):
    if df == 'neutral':
        return 0
    elif df == 'negative':
        return -1
    else:
        return 1
df_copy['target'] = df_copy.lables.apply(func)
df_copy.head()
```

```python
df_copy = df_copy.drop('lables', axis=1)
```

```python
df_copy = df_copy[df_copy.target != 0]
```

```python
X= df_copy['messages']
y=df_copy['target']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf
import ktrain
from ktrain import text


MODEL_NAME = 'albert-base-v2'
t = text.Transformer(MODEL_NAME, maxlen=400, classes=np.array([1,-1]))
trn = t.preprocess_train(np.array(X_train),np.array(y_train))
val = t.preprocess_test(np.array(X_test), np.array(y_test))
model = t.get_classifier()
learner = ktrain.get_learner(model, train_data=trn, val_data=val, batch_size=6)
```

```python
learner.fit_onecycle(2e-5, 1)
predictor = ktrain.get_predictor(learner.model,preproc = t)
```
