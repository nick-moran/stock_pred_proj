```python
articles = news_data['text']
punc_free = [s.translate(str.maketrans('', '', string.punctuation)) for s in articles]
punc_free = [re.sub(r'[0-9]+', '', s) for s in punc_free]
tokenized_articles = [word_tokenize(text) for text in punc_free]
tokenized_articles = [article for article in tokenized_articles]
filtered_articles = [[ps.stem(w.lower()) for w in tokenized if not w.lower() in stop_words] for tokenized in tokenized_articles]
```

```python
len(filtered_articles)
```

```python
st_test = ".".join(filtered_articles[0])
str_filt = [".".join(filtered_article) for filtered_article in filtered_articles]
news_data['text'] = str_filt

news_data = news_data.drop_duplicates(subset=['text'])

news_data.to_csv('./datasets/news_data_filtered_unique.csv',index=False)
```
